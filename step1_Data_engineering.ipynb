Given the data is recorded by research staff. We wanted to engineer some the metrics to our model's convinience. 

Convertions:
for a course of 36 months, food consumption records at 

*   0/never == 0
*   1/once a month == 36,
*   2/once a fortnight == 72
*   3/once a week == 144
*   4/2-3 times per week == 422
*   5/4-6 times per week == 720
*   6/Daily == 1080
*   7/more than once daily == 2160.

```python
# create a list of our conditions
conditions = [
    (food_allergens['4months_milk_consumption'] == 0),
    (food_allergens['4months_milk_consumption'] == 1),
    (food_allergens['4months_milk_consumption'] == 2),
    (food_allergens['4months_milk_consumption'] == 3),
    (food_allergens['4months_milk_consumption'] == 4),
    (food_allergens['4months_milk_consumption'] == 5),
    (food_allergens['4months_milk_consumption'] == 6),
    (food_allergens['4months_milk_consumption'] == 7)
    ]

# create a list of the values we want to assign for each condition
values = [0, 4, 8, 12, 32, 64, 122, 183]

# create a new column and use np.select to assign values to it using our lists as arguments
food_allergens['milk_baby_food'] = np.select(conditions, values)

# display updated DataFrame
food_allergens.head()
```

```python
lst = ['freq of milk baby food consumed at 36 months',
'freq of wheat baby food consumed at 36 months','freq of egg baby food consumed at 36 months',
'freq of cereal consumed at 36 months','freq of bread consumed at 36 months',
'freq of dried pasta consumed at 36 months','freq of fresh pasta consumed at 36 months',
'freq of pastry consumed at 36 months','freq of tortilla consumed at 36 months','freq of egg consumed at 36 months',
'freq of croissant consumed at 36 months','freq of scone consumed at 36 months','freq of cake consumed at 36 months',
'freq of pancake consumed at 36 months','freq of jaffa consumed at 36 months','freq of quiche consumed at 36 months',
'freq of mayo consumed at 36 months','freq of yoghurt consumed at 36 months','freq of custard consumed at 36 months',
'freq of cream consumed at 36 months','freq of soft cheese consumed at 36 months',
'freq of cheese sauce consumed at 36 months','freq of hard cheese consumed at 36 months',
'freq of fish fingers consumed at 36 months','freq of fish cakes consumed at 36 months',
'freq of fish pie consumed at 36 months','freq of fish baby food consumed at 36 months',
'freq of peanut butter consumed at 36 months','freq of peanut cereal consumed at 36 months',
'freq of peanut choc consumed at 36 months','freq of hummus consumed at 36 months',
'freq of tahini consumed at 36 months','freq of sesame crackers consumed at 36 months',
'freq of sesame snaps consumed at 36 months','freq of sesame bread consumed at 36 months']


for i in lst:
  conditions = [
    (food_allergens[i] == 0),
    (food_allergens[i] == 1),
    (food_allergens[i] == 2),
    (food_allergens[i] == 3),
    (food_allergens[i] == 4),
    (food_allergens[i] == 5),
    (food_allergens[i] == 6),
    (food_allergens[i] == 7)
    ]

  # create a list of the values we want to assign for each condition
  values = [0, 36, 72, 144, 422, 720, 1080, 2160]
  food_allergens[i+'_new']  = np.select(conditions, values)
```
